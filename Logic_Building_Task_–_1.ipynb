```python
# 1. User Login Check
username_input = input()
password_input = input()

username = "admin"
password = "1234"

if username_input == username and password_input == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
# 2. Pass / Fail Analyzer
marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("Pass Students:", pass_count)
print("Fail Students:", fail_count)
```

```python
# 3. Simple Data Cleaner
names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

for name in names:
    cleaned_names.append(name.strip().lower())

print(cleaned_names)
```

```python
# 4. Message Length Analyzer
messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
    length = len(message)
    print("Length:", length)

    if length > 10:
        print("Flagged: Message longer than 10 characters")
```

```python
# 5. Error Message Detector
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

for log in logs:
    if log == "ERROR":
        error_count += 1

print("Total Errors:", error_count)
```
