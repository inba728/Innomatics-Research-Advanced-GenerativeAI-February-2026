```python
# Problem Statement 1:Smart Parking Lot Management System
def manage_parking(vehicle_logs, capacity):
    parked = 0

    for log in vehicle_logs:
        if log == "IN":
            parked += 1
        elif log == "OUT":
            parked -= 1

    print("Currently Parked Vehicles:", parked)

    if parked > capacity:
        print("Parking Status: Full / Alert!")
    else:
        print("Parking Status: Available")

capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

manage_parking(vehicle_logs, capacity)
```

```python
# Problem Statement 2:Online Food Delivery Time Estimator
def estimate_delivery_time(distance, traffic, weather):
    # Base Travel: 3 mins per km
    base_time = distance * 3

    # Traffic Multipliers
    traffic_weights = {"Low": 1.0, "Moderate": 1.2, "High": 1.5}

    # Weather Delays (Flat minutes)
    weather_delays = {"Clear": 0, "Rainy": 10, "Stormy": 20}

    # Restaurant Prep Time
    prep_time = 9

    # Calculation
    total_eta = (base_time * traffic_weights.get(traffic, 1.0)) + weather_delays.get(weather, 0) + prep_time

    return f"Estimated Delivery Time: {round(total_eta)} minutes"

# Sample Input
print(estimate_delivery_time(8, "High", "Rainy"))
```

```python
#  Problem Statement 3:Movie Theatre Seat Occupancy Analyzer
def analyze_theatre_occupancy(total_seats, booked_seats):
    # Calculate occupancy
    booked_count = len(booked_seats)
    occupancy_percentage = (booked_count / total_seats) * 100

    # Determine Show Status
    if occupancy_percentage == 100:
        status = "Housefull"
    elif occupancy_percentage >= 75:
        status = "Almost Full"
    else:
        status = "Seats Available"

    # Final Output
    return f"Occupancy: {int(occupancy_percentage)}%\nShow Status: {status}"

# Sample Input
total = 200
booked = [1] * 150

print(analyze_theatre_occupancy(total, booked))
```

```python
#  Problem Statement 4:Cloud Server Load Classification System
def classify_server_load(cpu_readings):
    # Calculate average CPU load
    avg_cpu = sum(cpu_readings) / len(cpu_readings)

    # Classify Server Status
    if avg_cpu >= 80:
        status = "Critical"
    elif 50 <= avg_cpu < 80:
        status = "Warning"
    else:
        status = "Normal"

    return f"Average CPU Load: {int(avg_cpu)}%\nServer Status: {status}"

# Sample Input
cpu_readings = [45, 60, 70, 85, 90]

# Output
print(classify_server_load(cpu_readings))
```

```python
#  Problem Statement 5:Smart Classroom Resource Usage Monitor
def monitor_classroom_usage(resource_usage):
    # Define usage threshold (e.g., 8 hours)
    threshold = 8
    overused = []

    # Identify resources exceeding threshold
    for resource, hours in resource_usage.items():
        if hours >= threshold:
            overused.append(resource)

    # Determine alert status
    overused_str = ", ".join(overused) if overused else "None"
    alert = "Yes" if overused else "No"

    return f"Overused Resources: {overused_str}\nEnergy Alert: {alert}"

# Sample Input
usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

# Output
print(monitor_classroom_usage(usage))
```

```python
#  Problem Statement 6: Online Event Registration Capacity Controller
def manage_event_registrations(capacity, total_registrations):
    # Determine confirmed and waitlisted counts
    if total_registrations > capacity:
        confirmed = capacity
        waitlisted = total_registrations - capacity
        status = "Closed"
    else:
        confirmed = total_registrations
        waitlisted = 0
        status = "Open"

    return f"Confirmed Registrations: {confirmed}\nWaitlisted Users: {waitlisted}\nRegistration Status: {status}"

# Sample Input
event_capacity = 100
registrations = 105

# Output
print(manage_event_registrations(event_capacity, registrations))
```
